```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/home/oem/KYM/인공지능프로그래밍-송주환교수님/tbook.csv")
```

### 1. 출판년도가 1998인 년도인 목록 검색

```python
df[df["yea"] == '1998']
```

### 2. 출판년도가 1999년-2003년인 목록 검색

```python
df.info()
```

```python
# 'yea'열의 데이터 타입을 문자열에서 정수로 변경합니다.
df['yea'] = df['yea'].str.extract('(\d+)').astype(int)

# 'yea' 열에서 조건에 맞는 행을 추출합니다.
df[(df['yea'] >= 1999) & (df['yea'] <= 2003)]
```

### 3. 출판년도가 2002년 또는 2005년 인 목록 검색

```python
df[(df['yea'] == 2002) | (df['yea'] == 2005)]
```

### 4. 청구기호가 005로 시작하는 목록 검색

```python
df[df['symb'].str.startswith('005')]
```

### 5. 저자가 김씨인 목록 검색

```python
df[df['author'].str.startswith('김')]
```

### 6. 자료명에 데이터베이스가 포함된 목록 검색

```python
df[df['title'].str.contains('데이터베이스')]
```

### 7. 저자가 김씨 또는 이씨인 목록 검색

```python
df[df['author'].str.startswith(('김', '이'))]
```

### 8. 출판사가 길벗, 이한, 구민사, 정일 인 목록 검색

```python
df[df['publ'].isin(['길벗', '이한', '구민사', '정일'])]
```

### 9. 자료명에 데이터베이스가 포함되어 있지 않은 목록 검색

#### 참고로, '데이터베이스' 와 '데이터 베이스' 띄어쓰기 차이 존재함

```python
df[~df['title'].str.contains('데이터베이스')]
```

### 10. ISBN이 없는 목록 검색

```python
df[df['isbn'].isnull()]
```

### 11. 저자이름이 호로 끝나는 목록 검색

```python
df[df['author'].str.endswith('호')]
```

### 12. 출판년도가 2로 끝나는 목록검색

```python
df[df['yea'].astype(str).str.endswith('2')]
```
