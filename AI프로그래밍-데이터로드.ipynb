## 저울 1,2,3 으로 각각 추출

```python
data_jeoul1 = data[data[' jeoulID'] == 1]
data_jeoul2 = data[data[' jeoulID'] == 2]
data_jeoul3 = data[data[' jeoulID'] == 3]
```

```python
print(data_jeoul1[' getTime'])
```

## 각저울을 getTime 년-월-일 시:분:초 를 기준으로 시간순정렬함

```python
data_jeoul1[' getTime'] = pd.to_datetime(data_jeoul1[' getTime'])
data_jeoul2[' getTime'] = pd.to_datetime(data_jeoul2[' getTime'])
data_jeoul3[' getTime'] = pd.to_datetime(data_jeoul3[' getTime'])

# getTime 열을 기준으로 오름차순 정렬
data_jeoul1_sorted = data_jeoul1.sort_values(by=' getTime', ascending=True)
data_jeoul2_sorted = data_jeoul2.sort_values(by=' getTime', ascending=True)
data_jeoul3_sorted = data_jeoul3.sort_values(by=' getTime', ascending=True)

# 결과 출력
print(data_jeoul1_sorted)
```

```python
id_vars = ['_id', ' farmID', ' dongID', ' jeoulID', ' getTime', ' temp', ' humi', ' co', ' nh', ' dust']
value_vars = [' w{:02d}'.format(i) for i in range(1, 61)]
data_jeoul1_melted = pd.melt(data_jeoul1_sorted, id_vars=id_vars, value_vars=value_vars, var_name='variable', value_name='value')
print(data_jeoul1_melted)

id_vars = ['_id', ' farmID', ' dongID', ' jeoulID', ' getTime', ' temp', ' humi', ' co', ' nh', ' dust']
value_vars = [' w{:02d}'.format(i) for i in range(1, 61)]
data_jeoul2_melted = pd.melt(data_jeoul2_sorted, id_vars=id_vars, value_vars=value_vars, var_name='variable', value_name='value')


id_vars = ['_id', ' farmID', ' dongID', ' jeoulID', ' getTime', ' temp', ' humi', ' co', ' nh', ' dust']
value_vars = [' w{:02d}'.format(i) for i in range(1, 61)]
data_jeoul3_melted = pd.melt(data_jeoul3_sorted, id_vars=id_vars, value_vars=value_vars, var_name='variable', value_name='value')
```

```python
print(data_jeoul1_melted.info())
print(data_jeoul2_melted.info())
print(data_jeoul3_melted.info())
```
